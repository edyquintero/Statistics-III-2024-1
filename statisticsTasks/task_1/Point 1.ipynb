### 1.1 ¿What is the mean, median, and standard deviation?, and the mode and mode repetition values for categorical data.


#### 1.1.1 Mean

```python
meanCol1 = column1.sum()/len(column1)
meanCol2 = column2.sum()/len(column2)
meanCol3 = column3.sum()/len(column3)

print(f"Mean of PIB millions: {meanCol1}")
print(f"Mean of Population: {meanCol2}")
print(f"Mean of PIB per capita milions: {meanCol3}")
```

#### 1.1.2 Median

```python
def calculateMedian(column):
    sortedColumn = sorted(column)
    n = len(sortedColumn)
    middleIndex = n // 2
    if n % 2 == 0:
        median = (sortedColumn[middleIndex - 1] + sortedColumn[middleIndex]) / 2
    else:
        median = sortedColumn[middleIndex]
    return median

median1 = calculateMedian(column1)
median2 = calculateMedian(column2)
median3 = calculateMedian(column3)

print(f"Median of PIB millions: {median1}")
print(f"Median of Population: {median2}")
print(f"Median of PIB per capita Millions: {median3}")
```

#### 1.1.3 Standard deviation

```python
def calculateStandarDeviation(column, mean):
    n = len(column)
    squaredDiffs = [(x - mean) ** 2 for x in column]
    variance = sum(squaredDiffs) / n
    standarDeviation = variance ** 0.5
    return standarDeviation

standarDeviation1 = calculateStandarDeviation(column1, meanCol1)
standarDeviation2 = calculateStandarDeviation(column2, meanCol2)
standarDeviation3 = calculateStandarDeviation(column3, meanCol3)

print(f"Standard Deviation of PIB millions: {standarDeviation1}")
print(f"Standard Deviation of Population: {standarDeviation1}")
print(f"Standard Deviation of PIB per capita Millions: {standarDeviation1}")
```

#### 1.1.4 Mode and mode repetition values for categorical data

```python
modeValueCounts = {}
categoricalColumns = [column1, column2, column3]

for column in categoricalColumns:
    modeValueCounts[column.name] = {
        'mode': column.mode().tolist(),
        'valueCounts': column.value_counts().to_dict()
    }

print("Mode and value counts for categorical columns:")
print(f"Mode: {df.mode().iloc[0]}")
for column, values in modeValueCounts.items():
    print(f"Column: {column}")
    print(f"Mode: {values['mode']}")
    print(f"Mode value counts:")
    for value, counts in values['valueCounts'].items():
        print(f"{value}: {counts}")
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots as sp

def calculateQuartile(column, cuartile):
    column = sorted(column)
    columnLength = len(column)
    qIndex = cuartile * (columnLength // 4)
    return column[qIndex] if columnLength % 4 != 0 else (column[qIndex] + column[qIndex + 1]) / 2

def calculateInterquartileRange(q1, q3):
    return q3 - q1

def getExtremeValues(column):
    column = sorted(column)
    return [column[0], column[-1]]

q1x_1 = calculateQuartile(column1, 1) 
q3x_1 = calculateQuartile(column1, 3)
iqrx_1 = calculateInterquartileRange(q1x_1, q3x_1)
upperx_1 = q3x_1 + 1.5 * iqrx_1
lowerx_1 = q1x_1 - 1.5 * iqrx_1
extremeValues = getExtremeValues(column1)

q1x_2 = calculateQuartile(column2, 1)
q3x_2 = calculateQuartile(column2, 3)
iqrx_2 = calculateInterquartileRange(q1x_2, q3x_2)
upperx_2 = q3x_2 + 1.5 * iqrx_2
lowerx_2 = q1x_2 - 1.5 * iqrx_2
extremeValues = getExtremeValues(column2)

q1x_3 = calculateQuartile(column3, 1)
q3x_3 = calculateQuartile(column3, 3)
iqrx_3 = calculateInterquartileRange(q1x_3, q3x_3)
upperx_3 = q3x_3 + 1.5 * iqrx_3
lowerx_3 = q1x_3 - 1.5 * iqrx_3
extremeValues = getExtremeValues(column3)

fig, axs = plt.subplots(3, 2, figsize=(15, 18))

axs[0, 0].boxplot(column1)
axs[0, 0].set_title('Boxplot of PIB millions')
axs[0, 1].boxplot(column2)
axs[0, 1].set_title('Boxplot of Population')
axs[1, 0].boxplot(column3)
axs[1, 0].set_title('Boxplot of PIB per capita millions')

axs[1, 1].hist(column1, bins=20, color='skyblue', alpha=0.7)
axs[1, 1].set_title('Histogram of PIB millions')
axs[2, 0].hist(column2, bins=20, color='skyblue', alpha=0.7)
axs[2, 0].set_title('Histogram of Population')
axs[2, 1].hist(column3, bins=20, color='skyblue', alpha=0.7)
axs[2, 1].set_title('Histogram of PIB per capita millions')

plt.tight_layout()
plt.show()
```

### 1.3 What is the covariance between the 2 variables x1 (PIB millions) and x2 (Population)?

```python
def calculateCovariance(columnX, columnY):
    columnLenght = len(columnX)

    meanX = sum(columnX) / columnLenght
    meanY = sum(columnY) / columnLenght

    return sum((columnX[i] - meanX) * (columnY[i] - meanY) for i in range(columnLenght)) / columnLenght

covariance = calculateCovariance(column1, column2)

print(f"The covariance is equal to: {covariance}")
```

### 1.4 What is the correlation between the variable x1 (PIB millions) and x2 (Population) (Calculate by hand)?

```python
def calculateCorrelation(columnX, columnY):
    meanX = sum(columnX) / len(columnX)
    meanY = sum(columnY) / len(columnY)

    covXY = sum((x - meanX) * (y - meanY) for x, y in zip(columnX, columnY)) / len(columnX)
    varX = sum((x - meanX) ** 2 for x in columnX) / len(columnX)
    varY = sum((y - meanY) ** 2 for y in columnY) / len(columnY)

    correlation = covXY / (varX ** 0.5 * varY ** 0.5)
    
    return correlation

corrrelation = calculateCorrelation(column1, column2)
print(corrrelation)
```

### 1.5 Explain the relationship between covariance and correlation.

Covariance is a raw measure of the relationship between two variables, whereas correlation is a standardized measure indicating both the direction and strength of the relationship between the two variables. Correlation is more useful in practice because it provides a more interpretable and comparative measure of the relationship between variables.
